```python
import matplotlib.pyplot as plt
#1. Bar Chart: Number of Movies vs TV Showsplt.figure()
plt.figure(figsize=(8,4))
df['type'].value_counts().plot(kind='bar')
plt.title('Number of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Pie Chart: Distribution of Content Ratings
plt.figure (figsize=(8,4))
df['rating'].value_counts().plot(kind="pie",autopct='%1.1f%%')
plt.title("Distribution of Content Ratings")
plt.show()
```

```python
# 3. Line Chart: Number of Titles Added per Year

plt.figure()
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title('Number of Titles Added per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Histogram: Movie Duration in Minutes
movies = df[df['type'] == 'Movie']
movies['duration_minutes'] = movies['duration'].str.replace(' min', '').astype(float)
plt.figure()
movies['duration_minutes'].plot(kind='hist', bins=20)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Bar Chart: Top 10 Countries by Total Releases
plt.figure()
df['country'].dropna().str.split(', ').explode().value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries by Total Releases')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
# 6. Stacked Bar Chart: Genres per Content Type
df_exploded = df.assign(genres=df['listed_in'].str.split(', ')).explode('genres')
genre_type = df_exploded.groupby(['genres', 'type']).size().unstack(fill_value=0)
plt.figure()
genre_type.head(10).plot(kind='bar', stacked=True)
plt.title('Genres per Content Type')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
# 7. Bar Chart: Top 10 Directors with Most Titles
plt.figure()
df['director'].dropna().str.split(', ').explode().value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Directors with Most Titles')
plt.xlabel('Director')
plt.ylabel('Count')
plt.show()
```

```python
# 8. Horizontal Bar Chart: Most Popular Actors/Actresses
plt.figure()
df['cast'].dropna().str.split(', ').explode().value_counts().head(10).plot(kind='barh')
plt.title('Top 10 Actors/Actresses')
plt.xlabel('Count')
plt.ylabel('Actor/Actress')
plt.show()
```

```python
#9. Heatmap: Correlation Between Release Year and Ratings
rating_year = df.groupby(['release_year', 'rating']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))
plt.imshow(rating_year, aspect='auto')
plt.title('Heatmap: Release Year vs Rating')
plt.xlabel('Rating')
plt.ylabel('Release Year')
plt.colorbar(label='Count')
plt.show()
```

```python
# 10. Bar Chart: Number of Titles by Top 15 Countries
plt.figure()
df['country'].dropna().str.split(', ').explode().value_counts().head(15).plot(kind='bar')
plt.title('Top 15 Countries by Number of Titles')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# 11. Line Plot: Cumulative Count of Releases Over Time
yearly_counts = df['release_year'].value_counts().sort_index()
cumulative = yearly_counts.cumsum()
plt.figure()
plt.plot(cumulative.index, cumulative.values)
plt.title('Cumulative Releases Over Time')
plt.xlabel('Year')
plt.ylabel('Cumulative Count')
plt.show()

```

```python
# 12. Scatter Plot: Release Year vs. Duration (Movies Only)
plt.figure()
plt.scatter(movies['release_year'], movies['duration_minutes'])
plt.title('Release Year vs Movie Duration')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.show()
```
